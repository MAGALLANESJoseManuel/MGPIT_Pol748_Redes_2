```python
import networkx as nx
G = nx.read_adjlist("net.adjlist",create_using=nx.DiGraph)
```

```python
import re

mapping={node:re.sub("@","",node) for node in sorted(G)}
```

```python
G_new=nx.relabel_nodes(G, mapping)
```

```python
len(G_new.edges())
```

```python
type(G_new)
```

```python
nx.to_pandas_adjacency(G_new)
```

```python
import pandas as pd
nx.to_pandas_adjacency(G_new).to_csv("seattleTop.csv",index_label='name')
```
